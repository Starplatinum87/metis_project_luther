## Page Scraping Functions

```python
# Function to convert dollars strings to integers
def dollars_to_int(dollars):
    try:
        dollars = dollars.replace('$', '')
    except:
        pass
    try:
        dollars = dollars.replace(',', '')
    except:
        pass
    if '.' in dollars:
        try:
            dollars = float(dollars)
        except:
            pass
    else:
        try: 
            dollars = int(dollars)
        except:
            pass
    return dollars
```

```python
# Function for returning film release date 
def get_ReleaseDate(meta):
    filmReleaseDate = ''
    for i in meta:
        i = i.values()
        try:
            if 'NON_TV_RELEASE' in list(i)[1][0]:
                filmReleaseDate = list(i)[0][:10]
                filmReleaseDate = datetime.strptime(filmReleaseDate, '%Y-%m-%d')
            else:
                continue
        except:
            continue

    return filmReleaseDate

# Function for returning MOVIEmeter at film release
def get_ReleaseRank(meta):
    meta = driver.execute_script('return graphData')
    filmReleaseRank = None
    for i in meta:
        i = i.values()
        try:
            if 'NON_TV_RELEASE' in list(i)[1][0]:
                filmReleaseRank = list(i)[-1]
            else:
                continue
        except:
            continue

    return filmReleaseRank
```

```python
# Create new data frame for daily gross based on films df
dailyGrossColumns = []
for i in range(1, 61):
    dailyGrossColumns.append('day'+str(i))
newColumns = list(films.columns) + dailyGrossColumns
filmDailyGross_v3 = pd.DataFrame(columns=newColumns)

# Create new data frame for daily theater numbers based on films df
dailyTheaterColumns = []
for i in range(1, 61):
    dailyTheaterColumns.append('day'+str(i))
newColumns = list(films.columns) + dailyTheaterColumns
filmDailyTheaters_v3 = pd.DataFrame(columns=newColumns)

# Create new data frame for daily MOVIEmeter rank based on films df
dailyRankColumns = []
for i in range(1, 61):
    dailyRankColumns.append('day'+str(i))
newColumns = list(films.columns) + dailyRankColumns
filmDailyRank_v3 = pd.DataFrame(columns=newColumns)
```

```python
# Create empty data frame for main data scraping of each page  below
films_master_columns = list(films.columns)[0:4]
adds = (['Target', 'Budget', 'Opening', 'USGross', 'WWGross',
                       'IntlGross', 'Release_dt', 'Release_str','MOVIEmeter', 
                       'MMReleaseDate', 'Ratings', 'RatingsVotes', 'Articles', 
                       'OpeningProp','NAGrossProp', 'IntlGrossProp', 'WorldGrossProp'])
films_master_columns = films_master_columns + adds
films_master = pd.DataFrame(columns=films_master_columns)
```

## Open Main Pickle File

```python
# Open main pickle reference file containing film name, IMDb Pro URL, title ID & current MOVIEmeter
with open('active_pickles/films.pickle','rb') as read_file:
    films = pickle.load(read_file) 
```

## Main Web Scraping Script

```python
# Loop through pages and get data
for index, data in films.iterrows(): # Iterate through rows of source dataframe films.iterrows
    driver.get(data['URL']) # Set driver URL to the URL of the film in each row
    
    data_list = list(data.iloc[0:4]) 
    data_list.append('TARGET')  # Append first 4 columns in films df to the front of each row
    
    ### Get primary Box Office data ###
    
    # Get budget 
    try:
        budget = driver.find_elements_by_xpath('//*[@id="box_office_summary"]/div[2]/div[1]/div/div[2]')[0].text
        budget = dollars_to_int(budget)
    except:
        budget = np.nan
    data_list.append(budget)

    # Get opening weekend 
    try:
        opening = driver.find_elements_by_xpath('//*[@id="box_office_summary"]/div[2]/div[2]/div/div[2]')[0].text
        opening = dollars_to_int(opening)
    except:
        opening = np.nan
    data_list.append(opening)

    # Get gross (US & Canada)
    try:
        naGross = driver.find_elements_by_xpath('//*[@id="box_office_summary"]/div[2]/div[3]/div/div[2]')[0].text
        naGross = dollars_to_int(naGross)
    except:
        naGross = np.nan
    data_list.append(naGross)

    # Get gross (World)
    try:
        worldGross = driver.find_elements_by_xpath('//*[@id="box_office_summary"]/div[2]/div[4]/div/div[2]')[0].text
        worldGross = dollars_to_int(worldGross)
    except:
        worldGross = np.nan
    data_list.append(worldGross)

    # Get international gross
    try:
        intGross = worldGross - naGross
    except:
        intGross = np.nan
    data_list.append(intGross)

    # Get release date and set to datetime object and string
    try:
        releaseDate = driver.find_elements_by_xpath('//*[@id="a-page"]/div[3]/div/div/div/div/div[2]/div[2]/div/div[2]/div[3]/span[1]/a')[0].text
        releaseDate_dt = datetime.strptime(releaseDate.strip(), '%b %d, %Y')
        releaseDate_str = datetime.strftime(releaseDate_dt, '%Y-%m-%d')
    except:
        releaseDate_dt = np.nan
        releaseDate_str = ''
    data_list.append(releaseDate_dt)
    data_list.append(releaseDate_str)
    
    # Function to get the javascript list of all date/rank combinations in the MOVIEmeter list
    meta = driver.execute_script('return graphData')
        
    
    # Return release date and MOVIEmeter rank with respective functions
    ReleaseRank = get_ReleaseRank(meta) # References function defined above loopo
    data_list.append(ReleaseRank)
    ReleaseDate = get_ReleaseDate(meta) # References functions defined above loop
    data_list.append(ReleaseDate) 
    
    # Get user ratings
    try:
        rating = driver.find_elements_by_xpath('//*[@id="rating_breakdown"]/div[1]/div/div/div[1]/div/div/div[1]/span[2]')[0].text
        rating = dollars_to_int(rating)
    except:
        rating = np.nan
    data_list.append(rating)
    
    # Get number of user ratings votes
    try:
        rating_votes = driver.find_elements_by_xpath('//*[@id="rating_breakdown"]/div[1]/div/div/div[1]/div/div/div[1]/span[3]')[0].text
        rating_votes = rating_votes.replace(',', '')
        regex = re.compile("\s(\d+)\svotes")
        rating_votes = regex.findall(rating_votes)[0]
        rating_votes = dollars_to_int(rating_votes)
    except:
        rating_votes = np.nan
    data_list.append(rating_votes)

    # Get number of articles
    try:
        articles = driver.find_elements_by_xpath('//*[@id="popularity_widget"]/div[2]/div/div/div[2]/div[2]/div/div[2]/span/a')[0].text
        articles = dollars_to_int(articles)
    except:
        articles = np.nan
    data_list.append(articles)
    
    # Opening revenue proportions of budget
    try:
        openingProp = opening / budget
    except:
        openingProp = np.nan
    data_list.append(openingProp)
        
    # Ending North American Gross proportion of budget    
    try:
        naGrossProp = naGross / budget
    except:
        naGrossProp = np.nan
    data_list.append(naGrossProp)
    
    # Ending international gross proportion of budget
    try:
        intGrossProp = intGross / budget
    except:
        intGrossProp = np.nan
    data_list.append(intGrossProp)
    
    # Ending worldwide gross proportion of budget
    try:
        worldGrossProp = worldGross / budget
    except:
        worldGrossProp = np.nan
    data_list.append(worldGrossProp)
    
    ### APPEND DATA TO DATA FRAME ROW BY ROW###
    film_main_series = pd.Series(data_list, index=films_master.columns)
    films_master = films_master.append(film_main_series, ignore_index=True)
   
    
    ##########################
    # Additional Data Frames #
    ##########################
    
    # Wait intervals so the scraping can keep up
    # rand = random.uniform(1.0, 2.0)
    wait = 5 
    
    # Got to next page
    driver.get(data['URL'])
    # time.sleep(wait)
    
    # Go to Box Office tab
    try:
        boxOffice_element = driver.find_elements_by_xpath('//*[@id="const_tabs_heading"]/li[6]/a')[0]
        boxOffice_URL = boxOffice_element.get_attribute('href')
        driver.get(boxOffice_URL)
        time.sleep(wait)
    except:
        pass


    # Change view to daily numbers
    try: 
        select = Select(driver.find_element_by_xpath('//*[@id="chart_type_filter_dropdown"]'))
        select.select_by_visible_text('Daily')
        time.sleep(wait)
    except:
        pass
    
    # Get daily gross as series and concat to filmDailyGross dataframe
    dailyNumbers = list(data.iloc[0:4])
    for i in range(2, 62):  
        try:
            dailyGross = driver.find_elements_by_xpath('//*[@id="box_office_mojo"]/tbody/tr[' + str(i) + ']/td[2]/div/p[1]')[0].text
            dailyGross = dollars_to_int(dailyGross)
            dailyNumbers.append(dailyGross)
        except:
            dailyNumbers.append(np.nan)            
    # Create series from list and append series to filmDailyGross df
    dailyNumbersSeries = pd.Series(dailyNumbers, index=filmDailyGross_v3.columns)
    filmDailyGross_v3 = filmDailyGross_v3.append(dailyNumbersSeries, ignore_index=True)
    
    
    
    # Get daily theater numbers as series and concat to filmDailyTheaters
    dailyTheaterNumbers = list(data.iloc[0:4])
    for i in range(2, 62):  
        try:
            dailyTheaters = driver.find_elements_by_xpath('//*[@id="box_office_mojo"]/tbody/tr[' + str(i) + ']/td[5]/div')[0].text
            dailyTheaters = dollars_to_int(dailyTheaters)
            dailyTheaterNumbers.append(dailyTheaters)
        except:
            dailyTheaterNumbers.append(np.nan)
    # Create series from list and append series to filmDailyTheaters
    dailyTheatersSeries = pd.Series(dailyTheaterNumbers, index=filmDailyTheaters_v3.columns)
    filmDailyTheaters_v3 = filmDailyTheaters_v3.append(dailyTheatersSeries, ignore_index=True)
    
    
    
    # Get daily MOVIEmeter rank as series and concat to filmDailyRank
    dailyRankNumbers = list(data.iloc[0:4])
    for i in range(2, 62):  
        try:
            dailyRank = driver.find_elements_by_xpath('//*[@id="box_office_mojo"]/tbody/tr[' + str(i) + ']/td[2]/div/p[2]')[0].text
            dailyRank = dailyRank[6:]
            dailyRank = dollars_to_int(dailyRank)
            dailyRankNumbers.append(dailyRank)
        except:
            dailyRankNumbers.append(np.nan)
    # Create series from list and append series to filmDailyRank
    dailyRankSeries = pd.Series(dailyRankNumbers, index=filmDailyRank_v3.columns)
    filmDailyRank_v3 = filmDailyRank_v3.append(dailyRankSeries, ignore_index=True)

    

## Test output
#     print('Budget:', budget)
#     print('Opening:', opening)
#     print('NA Gross:', naGross)
#     print('Intl. Gross:', intGross)
#     print('World Gross:', worldGross)
#     print('Release Date (string):', releaseDate_str)
#     print('Release Date (datetime):', releaseDate_dt)
#     print('Opening Proportion:', openingProp)
#     print('NA Gross Proportion:', naGrossProp)
#     print('Intl. Gross Proportion:', intGrossProp)
#     print('World Gross Proportion:', worldGrossProp) 
#     print('News Articles:', articles)
#     print('MOVIEmeter Release Date:', ReleaseDate)
#     print('MOVIEmeter Rank:', ReleaseRank)
#     print('User Rating:', rating)
#     print('Votes:', rating_votes)
    
    # Wait interval to avoid suspicion
    # time.sleep(wait)
```
